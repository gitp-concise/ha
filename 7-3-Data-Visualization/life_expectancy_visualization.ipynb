```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Life Expectancy.csv')
df.head(20)
```

```python

# abs_sorted = df.corr()['Life expectancy'].abs().sort_values(ascending=True)
abs_sorted = df.select_dtypes(include=[int, float, bool]).corr()['Life expectancy '].abs().sort_values(ascending=True)
abs_sorted = abs_sorted[:-1]
fig, ax = plt.subplots()
ax = plt.barh(abs_sorted.index, abs_sorted)
plt.title('correlation of life expectancy and features')
plt.xlabel('features')
plt.ylabel('corr')
plt.show()
```
