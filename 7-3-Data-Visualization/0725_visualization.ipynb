## Visualizing Spotify Data

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
csv_path = 'spotify.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
# check for duplicate values
df.duplicated().value_counts()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.describe()
```

### 1번 시각화

```python
df_num = df.select_dtypes(include=[int, float, bool])

plt.figure(figsize=(15, 15))
sns.heatmap(df_num.corr(), annot=True, cmap="Reds")
plt.show()
```

```python
# [물음1] 개선점 제안
## 여기에 코드를 작성해주세요 ##
plt.figure(figsize=(15, 15))
sns.heatmap(df_num.corr(), annot=True, center=0, linewidths=1, linecolor="yellow", square=True)
plt.show()

```

```python
# total songs based on genres
fig = px.histogram(df.groupby('genre', as_index=False).count().sort_values(by='song', ascending=False),
                   x='genre', y='song', template='plotly_dark', title='<b>Total songs based on genres</b>')
fig.show()
```

```python
# popular genres based on popularity
fig = px.histogram(df.groupby('genre', as_index=False).sum().sort_values(by='popularity', ascending=False),
                   x='genre', y='popularity', color_discrete_sequence=['green'], template='plotly_dark',
                   title='<b>Popular genres based on popularity</b>')
fig.update_layout()
```

### 2번 시각화

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.scatterplot(x='instrumentalness', y='popularity',
                sizes=(1, 20),
                hue='duration_ms',
                size='explicit',
                linewidth=0.2,
                data=df, ax=ax, )

plt.title('Scatterplot of popularity')
plt.show()
```

```python
# [물음2] 개선점 제안
## 여기에 코드를 작성해주세요 ##
f, ax = plt.subplots(figsize=(14, 10))
sns.scatterplot(x='instrumentalness', y='popularity',
                sizes=(5, 40),
                hue='duration_ms',
                linewidth=0.2,
                data=df, ax=ax, size="duration_ms")

plt.xscale('log', base=10)
plt.title('Scatterplot of popularity')
plt.show()

```

```python
# [물음3] explicity가 popularity에 영향을 주는지 주지 않는지 판단하고 시각화를 통해 이를 정당화하세요.
## 여기에 코드를 작성
f, ax = plt.subplots(figsize=(14, 10))
sns.scatterplot(x='explicit', y='popularity',
                linewidth=0.2,
                data=df, ax=ax)

plt.title('Scatterplot of popularity')
plt.show()



```

```python
# [물음5] 적절한 시각화(들)
import plotly.graph_objects as go

df_num = df.select_dtypes(include=[int, float, bool])
top_10 = df_num.sort_values(by=["popularity"])[:10]
under_10 = df_num.sort_values(by=["popularity"])[10:-1]
top_10 = top_10.drop(columns=["year", "duration_ms", "popularity", "tempo"])
under_10 = under_10.drop(columns=["year", "duration_ms", "popularity", "tempo"])

go.Figure(
    data=[
        go.Scatterpolar(r=top_10.mean(), theta=top_10.columns.tolist()),
        go.Scatterpolar(r=under_10.mean(), theta=under_10.columns.tolist())
    ]
)

```
